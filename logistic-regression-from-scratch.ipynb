```python
df = pd.read_csv("/kaggle/input/weather-forecast-dataset/weather_forecast_data.csv")
df.head()
```

```python
y = df["Rain"].map({"rain": 1, "no rain": 0})
```

```python
x = df.drop("Rain",axis = 1)
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
def cost_function(X, y, w):
    z = np.dot(X, w)
    y_pred = 1 / (1 + np.exp(-z))
    m = len(y)
    cost = (-1 / m) * np.sum(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))

    return cost
```

```python
def get_gradient(X, y, w):
    m = len(y)
    z = np.dot(X, w)
    y_pred = 1 / (1 + np.exp(-z))
    grad = (1/m) * np.dot(X.T, (y_pred - y))
    return grad

    
```

```python
def gradient_descent(X, y, epochs, learning_rate):
    m, n = X.shape
    X = np.c_[np.ones((m,1)), X]
    w = np.zeros(n+1)
    
    for _ in range(epochs):
        w = w - learning_rate * get_gradient(X, y, w)
    
    return w


    
```

```python
w = gradient_descent(x,y,100,0.01)
```

```python
w
```

```python
def predict(x, w):
    m = len(x)
    x = np.c_[np.ones((m,1)), x]  
    z = np.dot(x, w)
    y_pred = 1 / (1 + np.exp(-z))
    return (y_pred >= 0.5).astype(int)
```

```python
y_predict = predict(x,w)
```

```python
accuracy = np.mean(y_predict == y)
print("Accuracy:", accuracy)
```
